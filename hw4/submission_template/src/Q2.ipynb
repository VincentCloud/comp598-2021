```python
from bokeh.io import show
from bokeh.layouts import column
from bokeh.models import CustomJS, Dropdown
```

```python
import pandas as pd
df = pd.read_csv('../data/filtered_data.csv')
```

```python
# drop the nan rows
df_without_na = df[df['Incident Zip'].notnull()]
df_without_na = df_without_na[df_without_na['Created Date'].notnull()]
df_without_na = df_without_na[df_without_na['Closed Date'].notnull()]
```

```python
df_without_na = df_without_na.drop('Unnamed: 0', 1)
```

```python
df_with_valid_timeline = df_without_na[pd.to_datetime(df_without_na['Created Date']) < pd.to_datetime(df_without_na['Closed Date'])]
```

```python
len(df_with_valid_timeline)
```

```python
menu = [(f'{zip}', f'{zip}') for zip in df_with_valid_timeline['Incident Zip'].unique().tolist()]

dropdown_zip1 = Dropdown(label="Zip Code 1", button_type="warning", menu=menu)
dropdown_zip1.js_on_event("menu_item_click", CustomJS(code="console.log('dropdown: ' + this.item, this.toString())"))

dropdown_zip2 = Dropdown(label="Zip Code 2", button_type="warning", menu=menu)
dropdown_zip2.js_on_event("menu_item_click", CustomJS(code="console.log('dropdown: ' + this.item, this.toString())"))
show(column(dropdown_zip1, dropdown_zip2))
```
