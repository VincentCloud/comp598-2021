# Data Collection

```python
import pandas as pd
import os
import numpy as np

url = 'https://github.com/fivethirtyeight/russian-troll-tweets/blob/master/IRAhandle_tweets_1.csv?raw=true'
df = pd.read_csv(url)
```

```python
num_of_tweets = 10000
df = df.head(num_of_tweets + 1)
```

```python
df
```

```python
# filter out the tweets that are in English and do not contain a question
english_df = df[df['language'] == 'English']
english_no_q_df = english_df[~english_df['content'].str.contains('\?')]
english_no_q_df.head()
```

```python
data_path = 'filtered_data.tsv'
english_no_q_df.to_csv(data_path, sep='\t', index=False)
```

# Data Annotation

```python
# a. add the trump_mention feature
trump_mention = np.where(english_no_q_df['content'].str.contains(r'\bTrump\b', case=True), 'T', 'F')
```

```python
# b. create a new version of the dataset
english_no_q_df['trump_mention'] = trump_mention
```

```python
annotated_df = english_no_q_df[['tweet_id', 'publish_date', 'content', 'trump_mention']]
annotated_data_path = 'dataset.tsv'
annotated_df.to_csv(annotated_data_path, sep='\t', index=False)
annotated_df.head()
```

# Analysis

```python
# a. Compute the statistic: % of tweets that mention Trump
trump_mentioned_percentage = len(annotated_df[annotated_df['trump_mention'] == 'T']) / len(english_no_q_df['trump_mention'])
trump_mentioned_percentage_rounded = '%.3f' % (trump_mentioned_percentage * 100)
print(f'Original percentage (%): {trump_mentioned_percentage * 100}')
print(f'Rounded percentage (%): {trump_mentioned_percentage_rounded}')
```

```python
# Export to result.tsv
result_data_path = 'results.tsv'
result_df = pd.DataFrame(data={'result': ['frac-trump-mentions'], 'value': [trump_mentioned_percentage_rounded]})
result_df.to_csv(result_data_path, sep='\t', index=False)
```
